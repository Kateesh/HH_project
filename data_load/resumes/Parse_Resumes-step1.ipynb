```python
df1 = pd.read_csv("resumes_load3.csv", encoding = 'utf-16')
```

```python
df2 = pd.read_csv("resumes_load4.csv")
```

```python
df2 = df2[['birth_date', 'gender', 'area', 'title', 'specialization',
       'education_level', 'education', 'language', 'experience', 'skill_set',
       'skills', 'salary.amount', 'salary.currency']]
```

```python
df3 = df1.append(df2)
```

```python
df4 = df3[df3['area']=='Москва']
```

```python
df5 = df4.drop_duplicates().reset_index(drop = True)
```

```python
len(df5)
```

```python
df5.head(3)
```

```python
df5['skills'] = df5['skills'].fillna(' ')
```

```python
df_obr = df5[['birth_date', 'gender', 'area', 'education',
              'title', 'education_level', 'salary.amount', 'salary.currency']]
```

```python
# обработка specialization, experience, skill_set
```

```python
for i in range(len(df_obr)):
    
    #specialization    
    spec_json = ast.literal_eval(df5.loc[i,'specialization'])    
    df_obr.loc[i, 'specialization'] = spec_json[0]['name']
    
    
    #skill_set
    l = ast.literal_eval(df5.loc[i,'skill_set'])
    ss = ','.join(l)
    df_obr.loc[i, 'skill_set'] = ss
    
    #skills
    df_obr.loc[i, 'skills_desc'] = ' '.join(df5.loc[i, 'skills'].splitlines())
    
    #experience
    parsed_json = ast.literal_eval(df5.loc[i,'experience'])
    
    duration = 0
    all_description = ''

    for j in range(len(parsed_json)):
        start_date = datetime.strptime(parsed_json[j]['start'][:10],  '%d-%m-%Y')

        if parsed_json[j]['end'] is None:
                end_date = datetime.strptime('20-08-2022',  '%d-%m-%Y')
        else:
                end_date = datetime.strptime(parsed_json[j]['end'][:10],  '%d-%m-%Y')

        cur_dur = (end_date - start_date).days
        duration = duration + cur_dur

        all_description = all_description + ' '.join(parsed_json[j]['description'].splitlines())

    df_obr.loc[i, 'exp_length'] = duration
    df_obr.loc[i, 'exp_description'] = all_description
```

```python
df_obr.to_csv('resumes_step1.csv', encoding = 'utf-16', index = False)
```
