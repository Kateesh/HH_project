# Шаг 1. Загрузим id вакансий и минимальную информацию по каждой

```python
# за итерацию будем выгружать по 50 вакансий из каждой отрасли
```

```python
df_step1 = pd.DataFrame()

for k in specs:
    
    params = {
        'area': 1, # Поиск ощуществляется по вакансиям города Москва
        'page': 34, # Индекс страницы поиска на HH
        'per_page': 50, # Кол-во вакансий на 1 странице
        'specialization': k
    }
     
     
    req = requests.get('https://api.hh.ru/vacancies', params) # Посылаем запрос к API
    
    #сохраняем результат запроса в датафрейм
    
    e=req.json()
    data = []
    data.append(e)
    
    
    if data[0]['items']:

        vacancy_details = data[0]['items'][0].keys()
        df = pd.DataFrame(columns= list(vacancy_details))
        ind = 0

        for i in range(len(data)):
            for j in range(len(data[i]['items'])):
                df.loc[ind] = data[i]['items'][j]
                ind+=1
    
    df_step1 = df_step1.append(df)
    
    print(k)
```

```python
df_step1.to_csv("vacancies_step1_load35.csv")
```

# Шаг 2. Загрузим полную информацию по каждой вакансии из шага 1

```python
params = {
        'area': 1, # Поиск ощуществляется по вакансиям города Москва
    }

df_step2 = pd.DataFrame()
i = 0

for vid in list(df_step1['id']):
    
    ss = 'https://api.hh.ru/vacancies/{0}'.format(vid)
    
    req = requests.get(ss, params) # Посылаем запрос к API    
    
    df_step2 = df_step2.append(pd.json_normalize(req.json()))
    
    i = i+1
    if i%50==0:
        print(i)
```
