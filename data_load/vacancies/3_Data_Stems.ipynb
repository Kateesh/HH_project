```python
import pandas as pd
import re
from nltk.stem.snowball import SnowballStemmer
```

```python
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('data_processed.csv', encoding = 'utf-16')
```

```python
df.head(3)
```

```python
stemmer = SnowballStemmer('russian')

def clear_txt(txt):
    txt = txt.lower()
    #txt = re.sub('[/+_!@#$A-Za-z0-9\n.,:()""«»;-]', ' ', txt)
    txt = re.sub('[/+_!@#$0-9\n.,:()""«»;-]', ' ', txt)
    new_txt = ''
    for t in txt.split(' '):
        if len(t) > 0:
            new_txt = new_txt + stemmer.stem(t) + ' '
    return new_txt[:-1]
```

```python
df['description'] = df['description'].fillna(' ')
```

```python
%%time

for i in range(len(df)):
    
    s1 = re.sub('[^A-Za-zА-Яа-я]+', ' ', df.loc[i,'description'])
    
    df.loc[i, 'description_stem'] = clear_txt(s1)
    
    if i%5000==0:
        print('обработано', i)
```
