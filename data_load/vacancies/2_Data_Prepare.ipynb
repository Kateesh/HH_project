```python
# объединим все выгрузки из шага 1, оставим только релевантные поля

d3 = pd.DataFrame()

for i in range(1,36):
    path = 'vacancies_step1_load{0}.csv'.format(i)
    d = pd.read_csv(path)
    d2 = d[['id','name', 'snippet', 'alternate_url']]
    
    d3 = d3.append(d2)
    
    print(path)
```

```python
# объединим все выгрузки из шага 2, оставим только релевантные поля

d4 = pd.DataFrame()

for i in range(1,36):
    path = 'vacancies_step2_load{0}.csv'.format(i)
    d = pd.read_csv(path)
    d2 = d[['id', 'description', 'key_skills', 'specializations', 'professional_roles', 
        'languages', 'salary.from', 'salary.to', 'salary.currency', 
        'salary.gross', 'experience.name', 'experience.id', 'schedule.name', 'employment.name', 'employer.name']]
    
    d4 = d4.append(d2)
    
    print(path)
```

```python
# исходные выгрузки step1 и step2 не загружены в git
```

```python
d4 = d4.drop_duplicates()
```

```python
d3 = d3.drop_duplicates()
```

```python
# объединим итоговые данные шага 1 и шага 2 в один файл

df = d3.merge(d4, on='id', how='inner')
```

```python
df = df.drop_duplicates()
```

```python
df_obr = df[['id', 'name', 'salary.from', 'salary.to', 'alternate_url',
       'salary.currency', 'salary.gross', 'experience.name', 'experience.id', 'schedule.name',
       'employment.name', 'employer.name']]
```

```python
# Обработаем поля в различных форматах для дальнейшего использования
# список полей: snippet, description, key_skills, prof_role, specializations
```

```python
for i in range(len(df_obr)):
    
    #snippet
    snippet_json = ast.literal_eval(df.loc[i,'snippet'])    
    df_obr.loc[i, 'requirement'] = snippet_json['requirement']
    
    
    #key_skills
    ss = ''
    l = []
    skills_json = ast.literal_eval(df.loc[i,'key_skills'])
    for j in range(len(skills_json)):
        l.append(skills_json[j]['name'])
    
    ss = ",".join(l)
        
    df_obr.loc[i, 'skills'] = ss
    
    #description
    raw_html = df.loc[i,'description']
    df_obr.loc[i,'description'] = BeautifulSoup(raw_html, "lxml").text
    
    #professional_roles
    prof_json = ast.literal_eval(df.loc[i,'professional_roles'])    
    df_obr.loc[i, 'prof_role'] = prof_json[0]['name']
    
    #specializations
    sss = ''
    l = []
    skills_json = ast.literal_eval(df.loc[i,'specializations'])
    for j in range(len(skills_json)):
        l.append(skills_json[j]['profarea_name'])
    
    sss = ";".join(set(l))
    df_obr.loc[i, 'specializations'] = sss
    
    if i%5000==0:
        print('обработано', i)
```
